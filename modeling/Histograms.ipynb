### Read and Merge

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import scipy.stats as st
import plotly.express as px

data = pd.read_csv('all_variables.csv')
data = data[data['Age'] != 8.5]
# display(data)
# data.head(3)
# print(data.columns)

figure, axes = plt.subplots(1, 2, sharex=True, figsize=(12, 4))
sns.histplot(ax=axes[0],data=data, x=data['Age'], hue=data['Gender'], multiple="stack" , legend = True, kde = False)
axes[0].set_title('Age distribution by Gender (Male/Female)', size = 14)
axes[0].set_xlabel("Age", size = 14)
axes[0].set_ylabel("Count", size = 14 )

sns.histplot(ax=axes[1],data=data, x=data['Age'], hue=data['Group'], multiple="stack", legend = True, kde = False)
axes[1].set_title('Age distribution by Group (FD/FI)', size = 14)
axes[1].set_xlabel("Age", size = 14)
axes[1].set_ylabel("Count", size = 14 )
```
