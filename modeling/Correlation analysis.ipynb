```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
data = pd.read_csv("all_variables.csv")
display(data)
```

```python
figure, axes = plt.subplots(1, 2, sharex=False, sharey= False, figsize=(12, 4))
figure.suptitle('Correlation analysis between morphometrical measures and GEFT scores')

x = data['GEFT']
y = data['lh_postcentral_thickness']
r,p = pearsonr(y, x)
axes[0].set_title("Correlation analysis (r = "+str(float('%.3f'%r))+"; p = "+str(float('%.3f'%p))+')', size = 10 )
sns.regplot(ax=axes[0],x=x, y=y,scatter_kws={'s':8},ci=95, marker="o", color=".5", line_kws=dict(color="green"))
axes[0].set_xlabel("GEFT Score", size = 9)
axes[0].set_ylabel("Left postcentral thickness (mm)", size = 9)

x = data['GEFT']
y = data['lh_rostralmiddlefrontal_thickness']
r,p = pearsonr(y, x)
axes[1].set_title("Correlation analysis (r = "+str(float('%.3f'%r))+"; p = "+str(float('%.3f'%p))+')', size = 10 )
sns.regplot(ax=axes[1],x=x, y=y,scatter_kws={'s':8},ci=95, marker="o", color=".5", line_kws=dict(color="green"))
axes[1].set_xlabel("GEFT Score", size = 9)
axes[1].set_ylabel("Left rostralmiddlefrontal thickness (mm)", size = 9)
```
