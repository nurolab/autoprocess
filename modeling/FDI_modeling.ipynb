```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = pd.read_csv('sample_data/selected_features.csv')
# display(data.head())

std_scaler = StandardScaler()
data[['geft','lh_rostralmiddlefrontal_thickness','lh_postcentral_thickness','lh_postcentral_volume','rh_postcentral_volume']] = std_scaler.fit_transform(data[['geft','lh_rostralmiddlefrontal_thickness','lh_postcentral_thickness','lh_postcentral_volume','rh_postcentral_volume']])
# display(data.head())
average = pd.DataFrame()
n = 100
for i in range(0,n):
    X = data.iloc[:, 5:7].values
    y = data.iloc[:, 3].values
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2)

    clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
    models,predictions = clf.fit(X_train, X_test, y_train, y_test)

    models = models.sort_values(by=["Model"], ascending=False)
    # display(models)
    if(i == 0):
        average = models
    # models.to_csv(index=True)
    if(i > 0):
        average = (average.stack()+models.stack())
        average = average.unstack()

# print("\n")

average = average/n
display(average.sort_values(by=["Accuracy"], ascending=False))
```
